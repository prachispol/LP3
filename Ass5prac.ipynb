```python
cur_x=2
max_iter=1000
prev_step_size=1
precision=0.000001
df=lambda X:2*(X+3)
rate=0.01

def gradient_de(cur_x,max_iter,prev_step_size,precision,df,rate):
    iter=0;
    while(prev_step_size>precision or iter<max_iter ):
        prev_x=cur_x
        cur_x=cur_x-rate*df(prev_x)
        prev_step_size=abs(prev_x-cur_x)
        print("iter",iter,"cur_x ",cur_x)
        iter+=1;
    print("minima at",cur_x)
        
        
   
    
```

```python
gradient_de(cur_x,max_iter,prev_step_size,precision,df,rate)
```
